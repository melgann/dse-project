```python


# Example: Load a GeoJSON file with MultiPolygons and a 'name' column
gdf = gpd.read_file('district_and_planning_area.geojson')

# Ensure the geometry column is set
gdf = gdf.set_geometry('geometry')

# Function to calculate centroids and create a table
def create_centroid_table(gdf):
    centroids = []
    
    for index, row in gdf.iterrows():
        name = row['planning_area']  # Replace 'name' with the actual column name for the polygon name
        multipolygon = row['geometry']  # Get the MultiPolygon geometry
        
        # Calculate the centroid
        centroid = multipolygon.centroid
        
        # Extract latitude and longitude
        longitude = centroid.x
        latitude = centroid.y
        
        # Append to the list
        centroids.append({
            'name': name,
            'latitude': latitude,
            'longitude': longitude
        })
    
    # Convert the list to a pandas DataFrame
    centroid_table = pd.DataFrame(centroids)
    return centroid_table

# Create the centroid table
centroid_table = create_centroid_table(gdf)

# Display the table
print(centroid_table)
```
