```python


# Example: Load a GeoJSON file with MultiPolygons and a 'name' column
gdf = gpd.read_file('district_and_planning_area.geojson')

# Ensure the geometry column is set
gdf = gdf.set_geometry('geometry')

# Function to calculate centroids and create a table
def create_centroid_table(gdf):
    centroids = []
    
    for index, row in gdf.iterrows():
        name = row['planning_area']  # Replace 'name' with the actual column name for the polygon name
        multipolygon = row['geometry']  # Get the MultiPolygon geometry
        
        # Calculate the centroid
        centroid = multipolygon.centroid
        
        # Extract latitude and longitude
        longitude = centroid.x
        latitude = centroid.y
        
        # Append to the list
        centroids.append({
            'name': name,
            'latitude': latitude,
            'longitude': longitude
        })
    
    # Convert the list to a pandas DataFrame
    centroid_table = pd.DataFrame(centroids)
    return centroid_table

# Create the centroid table
centroid_table = create_centroid_table(gdf)

# Display the table


centroid_table.loc[38] = ["Tengah", 1.3555, 103.7308]
#print(centroid_table)
centroid_table.loc[7] = ["Changi Bay", 1.345010, 103.983208]
print(centroid_table)
```

```python
#Creating a Table with each Planning Area to each Street 
def add_streets_per_planning_area(df, street_names):
    
    # Create new rows for each planning area
    new_rows = []
    for area in df['name']:
        for street in street_names:
            new_rows.append({
                'Planning_area': area,
                'Street_name': street,
            })
    
    # Create new dataframe from the rows
    new_df = pd.DataFrame(new_rows)
    
    # Combine with original data
    #expanded_df = pd.concat([df, new_df], ignore_index=True)
    
    return new_df

street_names = ending_coordinates['Street_name']  
centroid_table1_expanded = add_streets_per_planning_area(centroid_table, street_names)
```

```python
#Full Join to check for missing values 
check_missing_pt_values = pd.merge(
    df_public_transport_timings, 
    centroid_table1_expanded, 
    on=['Street_name', 'Planning_area'], 
    how='outer'
)

#Selecting Relevant Rows
check_missing_pt_values = check_missing_pt_values[['Street_name', 'Planning_area', 'Time_taken']]
rows_with_na = check_missing_pt_values[check_missing_pt_values.isna().any(axis=1)]
rows_with_na["Planning_area"].unique()
```
